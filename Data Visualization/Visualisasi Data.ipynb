```python
import pandas as pd

df = pd.read_csv("https://gist.githubusercontent.com/aishwarya8615/89d9f36fc014dea62487f7347864d16a/raw/8629d284e13976dcb13bb0b27043224b9266fffa/Life_Expectancy_Data.csv")
df.head()
```

### A. Grafik Batang (Bar Plot)

```python
#Dengan Matplotlib
continent_counts = df['Continent'].value_counts()

X = continent_counts.index
y = continent_counts

plt.figure(figsize=(15,10))
plt.bar(X,y)
plt.title("Jumlah Negara tiap Benua (Continents)", fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.barh(X,y)

plt.xlabel("Benua", fontsize=16)
plt.ylabel("Jumlah", fontsize=16)
plt.title("Jumlah Negara tiap Benua (Continents)", fontsize=16)
plt.show()
```

```python
sns.set_theme()
plt.figure(figsize=(15,10))
sns.countplot(data=df, x="Continent",hue='Status')

plt.title("Jumlah Negara tiap Benua (Continents)", fontsize=16)
plt.show()
```

### B. Grafik Garis

Grafik garis umumnya digunakan untuk melihat perkembangan dari suatu nilai. Grafik garis digunakan bersama dengan tipe data yang bersifat date

```python
dfFilter = df[df['Continent'] == 'Asia']
dfYear = dfFilter.groupby(['Year'])['Life_expectancy '].mean()
plt.figure(figsize=(15,10))

plt.plot(dfYear.index,dfYear,marker='o')
plt.xlabel("Tahun", fontsize=16)
plt.ylabel("Angka Harapan Hidup", fontsize=16)
plt.title("Perkembangan Angka Harapan Hidup", fontsize=16)
plt.show()
```

### Grafik Pie

Grafik Pie digunakan untuk membandingkan jumlah dari suatu nilai yang bersifat kategorical

```python
data_pie =df['Status'].value_counts().rename_axis('Status').reset_index(name='Id_count')
data_pie
```

```python
plt.figure(figsize=(10,10))
plt.pie(data_pie.Id_count, labels=data_pie["Status"], startangle=90,autopct='%.1f%%',explode = (0.1,0),shadow=True)
plt.title('Status Negara')
plt.show()
```

### Grafik Scatter

Grafik scatter biasa digunakan untuk mengbandingkan antara 2 atribut yang bertipe numerik

```python
plt.figure(figsize=(15,10))
plt.scatter(x= df['Life_expectancy '],y= df['GDP'])

plt.ylabel("GDP", fontsize=16)
plt.xlabel("Angka Harapan Hidup", fontsize=16)
plt.title("Perbandingan GDP dengan Angka Harapan Hidup", fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(18,10))
sns.scatterplot(data=df,x=df['Life_expectancy '],y=df['GDP'],hue="Status")

plt.ylabel("GDP", fontsize=16)
plt.xlabel("Angka Harapan Hidup", fontsize=16)
plt.title("Perbandingan GDP dengan Angka Harapan Hidup", fontsize=16)
plt.show()
```

### Boxplot

Boxplot digunakan untuk melihat penyebaran dari suatu data yang bertipe kategorikal terhadap atribut yang bertipe data numerik 
(Categorical VS Numerik)

```python
plt.figure(figsize=(10,10))
sns.boxplot(data=df,x='Status',y='Life_expectancy ')
plt.show()
```
