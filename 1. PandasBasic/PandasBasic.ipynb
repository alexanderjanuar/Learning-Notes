### Series
**Series** adalah objek 1-dimensi yang berisi sequence nilai dan berasosiasi dengan label data, yang disebut indeks. Untuk membuat sebuah Series, kita dapat membentuknya dari sebuah array dengan memanggil fungsi Series pada **pandas**.

```python
Series = pd.Series([1, 2, 3, 4, 5])

print(Series)
print(type(Series))
```

Series, pada dasarnya, adalah satu kolom dari DataFrame. Kita dapat menetapkan label baris ke Series dengan menggunakan parameter indeks. Namun, Seri tidak memiliki nama kolom, hanya memiliki satu nama keseluruhan:

```python
Values = [30, 35, 40,30,10,5]
Index = ['2015 Sales', '2016 Sales', '2017 Sales',"2015 Sales","2020 Sales","2025 Sales"]

pd.Series(Values, index=Index, name='Product A')
```

### DataFrame
DataFrame merupakan tabel data yang terdapat kolom dan baris, dimana nilai-nilai yang terdapat di dalamnya dapat berupa tipe berbeda seperti numeric, string, boolean, dll. DataFrame mirip dengan data 2-dimensi dengan adanya baris dan kolom. Selain itu, DataFrame bisa dikatakan gabungan dari dictionary objek Series yang memiliki indeks yang sama.

<span style='background :Yellow' > pd.DataFrame()</span>

Terdapat berbagai macam cara untuk membentuk objek DataFrame. 
Salah satu cara yang biasa dilakukan untuk membentuk objek DataFrame dengan menggunakan data masukan berupa dictionary.

```python
#Data
Nama = ["Alex","Aisyah","Rizky","Dendi","Bagus","Ajeng","Alyani","Vauwez","Kulus","Wendra"]
Jurusan = ["Informatika","Kimia","Sipil","Kimia","Informatika","Elektro","Industri","Sipil","Kimia","Informatika"]
IPK = [2.5,3.2,3.9,2.5,3.1,3.8,3.1,3.7,3.8,1.0]

df = pd.DataFrame(
   {
       "Nama":Nama,
       "Jurusan" : Jurusan,
       "IPK" : IPK
   } 
)
```

```python
df.head()
```

#### Indexing in pandas
Terdapat dua operator didalam pandas yang digunakan untuk melakukan indexing
- iloc
- loc

#### Index-based selection (iloc)
Memilih data berdasarkan posisi index dalam data

```python
df.head()
```

```python
df.iloc[2:5]
```

#### Label-based selection (loc)
Nilai indeks data berdasarkan nama kolom

```python
df.head()
```

```python
df.loc[0:5,"Nama"]
```

```python
df.loc[0:5,["Nama","IPK"]]
```

#### Manipulating the index
Label-based selection bergantung pada index dalam penggunaanya. Didalam **Pandas**, kita dapat memanipulasi index sesuai yang kita mau

<span style='background :Yellow' > df.set_index()</span>

```python
df.set_index("Nama")
```

#### Conditional Statement
Kita juga dapat mengambil suatu baris didalam dataframe berdasarkan kondisi yang kita inginkan

```python
df.head
```

Siapa saja mahasiswa jurusan *informatika* dan memiliki IPK lebih dari *3.0*?

```python
df.loc[(df["Jurusan"] == "Informatika") & (df["IPK"] >= 3.0)]
```

Siapa saja mahasiswa jurusan informatika atau mahasiswa jurusan kimia?

```python
df.loc[(df["Jurusan"] == "Informatika") | (df["Jurusan"] == "Kimia")]
```

Siapa saja mahasiswa jurusan informatika,industri,kimia?

```python
Jurusan = ["Informatika","Industri"]
df.loc[df["Jurusan"].isin(Jurusan)]
```

## Grouping and Sorting


### Groupwise analysis

Dalam analisis data ada kalanya kita ingin melakukan agregasi data seperti mencari jumlah data, mencari rata-rata atau total nilai

Kita bisa gunakan fungsi GroupBy()

```python
dfLife.head(5)
```

```python
dfLife[dfLife["Country"] == "Indonesia"].head()
```

```python
dfLife.groupby("Country").min().head()
```

```python
dfLife.groupby("Country")[["GDP","Alcohol"]].mean().head().reset_index()
```

```python
dfLife.groupby("Country")[["GDP","Alcohol"]].agg([min,max]).head()
```

```python
dfLife.groupby(["Country","Year"])[["GDP"]].agg([min, max]).head().reset_index()
```
